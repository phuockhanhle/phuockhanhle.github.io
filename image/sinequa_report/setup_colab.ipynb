```python
import logging
from contextlib import nullcontext
import os
import sys
import time
# from torch.utils.data import DistributedSampler
import argparse
import mlflow
import pytorch_lightning as pl
from pytorch_lightning.callbacks import ModelCheckpoint
import torch

from KeywordExtraction_lightining import config
from KeywordExtraction_lightining import utils
from KeywordExtraction_lightining.bertkpe import tokenizer_class, dataloader
from KeywordExtraction_lightining import pl_model

logger = logging.getLogger()

MLFLOW_EXPERIMENT_NAME = "KPE"
```

```python
class Configure():
  def __init__(self):
    self.run_mode = True
    self.mlflow_run_name = "kpe1"
    self.max_train_steps = 10000 
    self.model_class = "bert2joint"
    self.dataset_class = "pubmed"
    self.pretrain_model_type = "bert-base-cased"
    self.max_train_epochs = 2
    self.train_batch_size = 2
    self.gradient_accumulation_steps = 1
    self.test_batch_size = 2
    self.preprocess_folder = "/tmp/data/prepro_dataset"
    self.pretrain_model_path = "/tmp/data/pretrain_model"
    self.cached_features_dir = "/tmp/data/cached_dir"
    self.display_iter = 10
    self.save_checkpoint = True
    self.use_viso = True
    self.learning_rate = 0.00005
    self.warmup_proportion = 0.1
    self.weight_decay =  0.01
    self.max_token = 510
    self.adam_epsilon = 0.00000001
    self.max_grad_norm = 1.0
    self.max_phrase_words = 5
    self.save_path = "/tmp"
    self.n_cores = 8
    self.data_workers = 1
    self.save_pytorch_model = True
    self.ditributed = True

    self.n_gpu = self.n_cores
    self.local_rank = -1
    self.device = None
    self.cuda = None
    self.checkpoint_file = ''

    self.cache_dir = os.path.join(self.pretrain_model_path, self.pretrain_model_type)
    self.preprocess_folder = os.path.join(self.preprocess_folder, self.dataset_class)

    folder_name = "%s_%s_%s_%s_" % (self.run_mode, self.model_class, self.dataset_class,
                                    self.pretrain_model_type.split('-')[0]) + time.strftime("%m.%d_%H.%M")
    self.save_folder = os.path.join(self.save_path, folder_name)

    # principle folder
    if not os.path.exists(self.save_path):
        os.mkdir(self.save_path)
    if not os.path.exists(self.save_folder):
        os.mkdir(self.save_folder)

    # viso folder
    if self.use_viso:
        self.viso_folder = os.path.join(self.save_folder, 'viso')
        if not os.path.exists(self.viso_folder):
            os.mkdir(self.viso_folder)

    # checkpoint folder
    if self.run_mode == 'train':
        self.checkpoint_folder = os.path.join(self.save_folder, 'checkpoints')
        if not os.path.exists(self.checkpoint_folder):
            os.mkdir(self.checkpoint_folder)

    elif self.run_mode == 'test':
        self.pred_folder = os.path.join(self.save_folder, 'predictions')
        if not os.path.exists(self.pred_folder):
            os.mkdir(self.pred_folder)

    # logging file
    self.log_file = os.path.join(self.save_folder, 'logging.txt')
    logger = logging.getLogger()
    logger.setLevel(logging.INFO)  # logger.setLevel(logging.DEBUG)
    fmt = logging.Formatter('%(asctime)s: [ %(message)s ]', '%m/%d/%Y %I:%M:%S %p')

    console = logging.StreamHandler()
    console.setFormatter(fmt)
    logger.addHandler(console)
    if self.log_file:
        logfile = logging.FileHandler(self.log_file, 'w')
        logfile.setFormatter(fmt)
        logger.addHandler(logfile)
    logger.info('COMMAND: %s' % ' '.join(sys.argv))
    logger.info("preprocess_folder = {}".format(self.preprocess_folder))
    logger.info("Pretrain Model Type = {}".format(self.pretrain_model_type)),

args = Configure()
```

```python
import KeywordExtraction_lightining.pl_data_config as pl_data_config

pl_data_config.save_data_config("/tmp/data", "/tmp/data")
```

```python
tokenizer = tokenizer_class[args.pretrain_model_type].from_pretrained(args.cache_dir)
data_train_config = pl_data_config.load_data_config("/tmp/data/config.json",
                                                     args.dataset_class, "train")
data_dev_config = pl_data_config.load_data_config("/tmp/data/config.json", 
                                                  args.dataset_class, "dev")
```

```python
def train(
        trainer_kwargs,
        data_train_config_,
        data_dev_config_,
        tokenizer_,
        seed,
):
    logging.info("Start training KPE model.")

    if args.mlflow_run_name:
        mlflow.set_experiment(MLFLOW_EXPERIMENT_NAME)

    if seed:
        pl.seed_everything(seed=seed, workers=True)

    logger.info("Instantiate KPE model.")
    model = pl_model.KeyphraseExtraction(
        data_train_config=data_train_config_,
        data_dev_config=data_dev_config_,
        tokenizer=tokenizer_,
        config=args,
    )

    logger.info("Instantiate call back")
    ckpt_callback = ModelCheckpoint(
        monitor="KPE call back",
        dirpath=args.save_folder,
        filename='{epoch}',
        mode="max",
        save_top_k=1,
        verbose=True,
    )

    trainer = pl.Trainer(
        callbacks=[ckpt_callback],
        default_root_dir=args.save_folder,
        num_sanity_val_steps=0,
        accumulate_grad_batches=args.gradient_accumulation_steps,
        gradient_clip_val=args.max_grad_norm,
        gradient_clip_algorithm="value",
        **trainer_kwargs,
    )

    logging.info("Train model.")
    with mlflow.start_run(run_name=args.mlflow_run_name) if args.mlflow_run_name else nullcontext():
        if args.mlflow_run_name:
            list_params_concern = ['max_train_steps', 'per_gpu_train_batch_size',
                                   'model_class', 'dataset_class', 'pretrain_model_type',
                                   'train_batch_size', 'gradient_accumulation_steps',
                                   'test_batch_size', 'warmup_proportion', 'weight_decay',
                                   'max_token', 'adam_epsilon', 'max_grad_norm', 'max_phrase_words',
                                   'learning_rate', 'max_train_epochs']
            for key, value in vars(args).items():
                if key in list_params_concern:
                    mlflow.log_param(key, value)

            trainer.fit(model=model)

        if args.save_pytorch_model:
            pt_model_path = os.path.join(args.save_folder, "model.pt")
            logging.info(f"Save pytorch model to '{pt_model_path}'.")
            torch.save(
                obj=model.model.state_dict(),
                f=pt_model_path,
            )
            if args.mlflow_run_name:
                logging.info("Log pytorch model as artifact to mlflow.")
                mlflow.log_artifact(pt_model_path)

    logger.info("Terminated successfully.")
```

```python
TRAINER_KWARGS = {
    "tpu_cores": args.n_cores,
    # "gpus": 1,
    "max_epochs": args.max_train_epochs,
    "limit_train_batches": 1.0,
    "log_every_n_steps": 1,
    %=20
}

train(
    TRAINER_KWARGS,
    data_train_config,
    data_dev_config,
    tokenizer,
    13,
)
```
