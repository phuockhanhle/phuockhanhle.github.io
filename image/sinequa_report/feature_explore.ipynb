```python
import json

def read_line_json(filename, nb_lines = 4):
  with open(filename, 'r') as json_file:
    json_list = list(json_file)

  for i in range(nb_lines):
    json_str = json_list[i]
    result = json.loads(json_str)
    print(f"result: {result}")
    print(isinstance(result, dict))
```

## dataset openkp

```python
read_line_json('./OpenKPDev.jsonl')
```

feature (raw) : url (string - link to documents) text(string - documents) VDOM (list of dictionnaty (id : index, text : keypharase, feature - *not knowing*, start_idx + end_idx : position))

```python
read_line_json('./openkp.dev.json')
```

feature (prepo) : url (string - link to documents), doc_words (list string - documents) word2block (segment of a phrase) bloc_features(list of list float - 
*not knowing*) keyphrases (list of list string, each list string is keypharas) start_end_pos (list of interval), present_keypharses (list of list string - *not knowing*)
